# **如何正则某一类字符呢**

+修饰的是前面的那个\d，比如说如果是5+**，那么就是可以有多个5，而？表示0个或1个都算，比如身，包括0个数字，而*表示0个或者多个，要么没有，要么就一起，将数据按照{}里的个数进行打包

```python
import re
text='身高：178，体重：168，学号：123456，密码：9527,123456，1，2，3'
target="123456"

print(re.findall(r'\d+',text))
print(re.findall(r'\d?',text))
print(re.findall(r'\d*',text))
print(re.findall(r'\d{3}',text))
```

d{3,4}表示3个或4个，-为中间的间隔表达

```python
text="mickes number is 18812345678,and another number is 198-8764233,his favorite number is 01234567891,and his zuoji number is 0571-52152166"
print(re.findall(r'\d{4}-\d{8}',text))
print(re.findall(r'\d{3,4}-\d{7,8}',text))
```

# |d代表着或的意思，用于处理多种情况的

```python
text="mickes number is 18812345678,and another number is 198-8764233,his favorite number is 01234567891,and his zuoji number is 0571-52152166"
print(re.findall(r'\d{4}-\d{8}',text))
print(re.findall(r'\d{3,4}-\d{7,8}|\d{10}',text))
```

^表示只取开头的,$表示取最后的,^放在最前，$放在句末

```python
text="18812345678mickes number is ,and another number is 198-8764233,his favorite number is 01234567891,and his zuoji number is 0571-52152166"
print(re.findall(r'^\d{4}-\d{8}',text))
print(re.findall(r'^\d{3,4}-\d{7,8}|^\d{10}',text))
print(re.findall(r'\d{3,4}-\d{7,8}$',text))
```

内部约束\w{3}表示3个字符，放在小括号(\w{3})中就成为一个分组，后面的（\1）表示引用，里面的1它里面的内容和第一个括号里的必须相同

```python
text='barbar carcar harhel har'
print(re.findall(r'(\w{3})|(\w{2})(\2)',text))
```
